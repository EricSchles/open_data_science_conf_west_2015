```python
import numpy as np
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.graphics.api import qqplot

dta = pd.read_csv("homeless_data.csv")
dta.index = pd.Index(sm.tsa.datetools.dates_from_range("2000m1","2015m4"))
#del dta["YEAR"]
dta.plot(figsize=(12,8))
plt.show()
```

```python
sm.tsa.datetools.dates_from_range("1983m1","2015m7")
```

```python
adults_in_families = dta["adults_in_families"]
adults_in_families.index = dta.index

fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(adults_in_families.values.squeeze(), lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(adults_in_families, lags=40, ax=ax2)
plt.show()
```

```python
arma_mod20 = sm.tsa.ARMA(adults_in_families, (2,0)).fit()
arma_mod30 = sm.tsa.ARMA(adults_in_families, (3,0)).fit()

```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)
ax = arma_mod30.resid.plot(ax=ax)
resid = arma_mod30.resid
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)
fig = qqplot(resid, line='q', ax=ax, fit=True)

```

```python
r,q,p = sm.tsa.acf(resid.values.squeeze(), qstat=True)
data = np.c_[range(1,41), r[1:], q, p]
table = pd.DataFrame(data, columns=['lag',"AC", "Q", "Prob(>Q)"])
print table.set_index('lag')

predict_sunspots = arma_mod30.predict("2012m1","2015m4", dynamic=True)
ax = dta.ix['1950':].plot(figsize=(12,8))
ax = predict_sunspots.plot(ax=ax, style="r--", label="Dynamic Prediction")
ax.legend()
ax.axis((-20.0,38.0,-4.0,200.0))
plt.show()
```
